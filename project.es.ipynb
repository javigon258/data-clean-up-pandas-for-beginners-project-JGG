#### Ejercicio 03. ¿Cuál es la casa más grande y la más pequeña del dataset? (★☆☆)

Imprime la dirección y el área de las casas seleccionadas. Por ejemplo:

`La casa más grande está ubicada en Calle Gran Vía, Nº38 y su superficie es de 5000 metros`

Este ejercicio es similar al anterior, pero buscamos las casas más grandes y pequeñas en base a su superficie.

```python
# TODO
casa_mas_grande = ds.loc[ds['surface'].idxmax(), ['id_realEstates', 'surface', 'realEstate_name']]
casa_mas_pequeña = ds.loc[ds['surface'].idxmin(), ['id_realEstates', 'surface', 'realEstate_name']]

print(f'La casa está en {casa_mas_grande.realEstate_name}, Nº {casa_mas_grande.id_realEstates} y su superficio es {casa_mas_grande.surface} metros')
print(f'La casa está en {casa_mas_pequeña.realEstate_name}, Nº {casa_mas_pequeña.id_realEstates} y su superficio es {casa_mas_pequeña.surface} metros')
```

#### Ejercicio 04. ¿Cuantas poblaciones contiene el dataset? (★☆☆)

#### Ejercicio 05. ¿El dataset contiene valores nulos (NAs)? (★☆☆)

Imprima un booleano (`True` o `False`) para verificar si hay valores nulos seguido de las columnas que contiene el NAs.

```python
# TODO
tiene_nulos = ds.isnull().any().any()

# Obtener las columnas que contienen valores nulos
columnas_con_nulos = ds.columns[ds.isnull().any()].tolist()

# Imprimir resultados
print(f"¿Hay valores nulos? {tiene_nulos}")
print(f"Columnas con valores nulos: {', '.join(columnas_con_nulos) if tiene_nulos else 'Ninguna'}")
```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido de la columna level5.

```python
# TODO
poblacion_filtrar = "Arroyomolinos (Madrid)"  # Reemplaza con el nombre real

# Filtrar el DataFrame para obtener solo los valores de esa población
df_filtrado = ds[ds['level5'] == poblacion_filtrar]
precios = df_filtrado['price']
precio_medio= precios.mean().round(2)
print(f'El precio medio en {poblacion_filtrar} es {precio_medio} €')
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt
import numpy as np

poblacion_filtrar = "Arroyomolinos (Madrid)"  

df_filtrado = ds[ds['level5'] == poblacion_filtrar]
precios = df_filtrado['price']

plt.figure(figsize = (10, 5))

plt.hist(precios, bins = 30, alpha = 0.7)

plt.title("Histograma")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```
